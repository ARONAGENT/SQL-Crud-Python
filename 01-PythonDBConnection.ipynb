# Python DB Connections process
1. **Installing PyMySQL**:
   - The first block shows the command `pip install pymysql` to install the PyMySQL library. This allows Python to communicate with MySQL databases.

2. **Importing PyMySQL**:
   - The next block imports the installed library using `import pymysql`.

3. **Connecting to the MySQL Database**:
   - The connection to the MySQL database is established using `pymysql.connect()`. The parameters include:
     - `host`: The URL of the MySQL database (`rohan-mysql-db-rohanbdconnections.c.aivencloud.com`).
     - `user`: Username (`avanadmin`).
     - `password`: Password for authentication.
     - `port`: Port number (19414).
     - `database`: The name of the database (`rohanbd`).
   - A success message `"connect with database successfully"` is printed upon successful connection.

4. **Creating a Cursor**:
   - A cursor object `curs` is created using `con.cursor()`. Cursors allow executing SQL queries from Python.

5. **Executing an SQL Query**:
   - The cursor executes a query to select all records from the `accounts` table with `curs.execute("select * from accounts")`.
   - The query returns `20`, indicating the number of rows returned from the table.

6. **Fetching and Printing Data**:
   - The data from the query is fetched using `curs.fetchall()`, and a loop is used to print each row of the fetched data.
