# Call transaction procedure created in database-
Delimiter //

create procedure transaction(in facc int,in tacc int,in amount float)

begin

update accounts set balance=balance-amount where accno=facc;

update accounts set balance=balance+amount where accno=tacc;
end//

delimiter ;


```python
facc=int(input('Enter from account no :'))
tacc=int(input('Enter to account no :'))
amount=float(input('Enter transfer amount :'))
```

```python
curs=con.cursor()
```

```python
curs.execute('call transaction(%d,%d,%f)'%(facc,tacc,amount))
```

```python
con.commit()
print('procedure call successfully')
```
